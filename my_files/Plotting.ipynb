### Slope intercept

y = mx + b

m = slope

b = intercept

```python
x = [-5,5]
m = 2
b = 1

y = [0, 0]
for i in range(0, len(x)):
    y[i] = m*x[i] + b
    

print(x,y)
plt.plot(x,y)

axis = plt.gca()
plt.axis("square")

plt.axhline(0, color="k")
plt.axvline(0, color="k")

axis.set_ylim([-5,5])
axis.set_xlim([-5,5])

plt.grid()
plt.show()
```

### using numpy

```python
x = [-5, 5]
m = [.7, -1.25]
b = [-2, .75]

for i in range(len(m)):
    y = m[i] * np.array(x) +b[i]
    plt.plot(x,y, label="y = %sx +%s" %(m[i], b[i]))

axis = plt.gca()
plt.axis("square")

plt.axhline(0, color="k")
plt.axvline(0, color="k")

axis.set_ylim([-5,5])
axis.set_xlim([-5,5])

plt.grid()
plt.show()
```

### Plotting with Sympy

```python
x = sym.symbols("x")
y = x**2
p = sym.plotting.plot(y, show=False)
p.ylim = [0,50]
p.xlim = [0,25]
p[0].line_color = "m"
p.title = "This is a nice plot"
p.show()
```

```python
x, a = sym.symbols("x, a")
expr = a/x
n, n2 = 1,2

p = symplot(expr.subs(a,n),  (x,-5,5), show=False)
p2 = symplot(expr.subs(a,n2),(x,-5,5), show=False)
p.extend(p2)

p.ylim = [-5,5]
p[0].label = '$y=%s$' % sym.latex(expr.subs(a,n))
p[1].label = '$y=%s$' % sym.latex(expr.subs(a,n2))

p[0].line_color = "r"

p.legend = True
p.show()
```

### Making images from matrices

```python
A = [[1,2], [1,4]]
A = np.array(A)
plt.imshow(A)
plt.xticks([0,1])
plt.yticks([0,1])
plt.show()
```

```python
A = np.zeros((10,14))

for i in range(0, A.shape[0]):
    for j in range(0, A.shape[1]):
        A[i,j] = 3*i - 4*j

fig, ax = plt.subplots(figsize=(14,7))
ax.imshow(A)
for i in range(0, A.shape[0]):
    for j in range(0, A.shape[1]):
        plt.text(j,i, int(A[i,j]), horizontalalignment="center",
                verticalalignment="center")

plt.set_cmap("twilight_shifted")
plt.show()
```

### Images from Matrices - making a chess board

```python
(-1)**8
```

```python
n = 20
A = np.zeros([n,n])

for i in range(0, A.shape[0]):
    for j in range(0, A.shape[1]):
        A[i,j] = (-1)**(i+j)

fig,ax = plt.subplots(figsize=(14,7))
plt.set_cmap("gray")
plt.tick_params(labelleft=False, labelbottom=False)
ax.imshow(A)
plt.show()
```

### Drawing patches with polygons

```python
y = np.array([ [1,1], [2,3], [3,1]])
y1 = np.array([ [2,3], [3,5], [3,1] ])
y2 = np.array([ [2,3], [1,5], [1,1] ])
y3 = np.array([ [2,3], [1,5], [3,5] ])
p = Polygon(y, facecolor="m", alpha=0.5)
p1 = Polygon(y1, facecolor="b", alpha=0.5)
p2 = Polygon(y2, facecolor="b", alpha=0.5)
p3 = Polygon(y3, facecolor="m", alpha=0.5)


fig, ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)
ax.add_patch(p2)
ax.add_patch(p3)
ax.set_xlim([0,6])
ax.set_ylim([0,6])
plt.show()
```

```python
x = np.linspace(-2,2, 101)
f = -x**2

pnts = np.vstack((x,f)).T
p = Polygon(pnts, facecolor="g", edgecolor="k", alpha=.2)

s = np.array([ [-.5, -4], [-.5,-2.5], [.5,-2.5] , [.5,-4]])
p1 = Polygon(s, facecolor="k")

fig,ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)

plt.plot(x,f, "k")
plt.plot([-2,2], [-4,-4], "k")
plt.axis("off")
plt.show()
```
